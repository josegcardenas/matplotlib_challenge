While testing treatments, we can appreciate the equal division of male and female specimens because treatments would incorporate men and women in future developments. We generated a line plot to compare the effectiveness of the Capomulin regimen against other similar drug regimens.
####  When we plot the average weight of a mouse and its corresponding tumor volume, we can see a strong positive correlation that determines an increase in weight will increase tumor volume. Initially, when looking through mouse ID=r157’s data there was no effect of the treatments because the tumor volume continued growing. However, given the time frame which the mouse had treatments, we can safely assume it did not have sufficient treatments to see a change or was too far gone for the treatments to take effect and we can determine it died early. Upon using another mouse, ID= s185, we see a much different story. Capomulin was very effective on this mouse decreasing tumor volume over time.

Although, Capomulin had great results it seems Ramicane performed slightly better with lower tumor volumes and for this reason we should continue testing with these two regimens.
```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_df = pd.merge(mouse_metadata, study_results, how = "outer", on = "Mouse ID")
# Display the data table for preview
combined_df.head()
```

```python
# Checking the number of mice.
number_of_mice = combined_df["Mouse ID"].unique()
print(f" There is {len(number_of_mice)} mice!")
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice_df = combined_df[combined_df.duplicated(['Mouse ID', 'Timepoint'])]
duplicate_mice_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_df = combined_df[combined_df.duplicated(['Mouse ID'])]
duplicate_mouse_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = combined_df [combined_df["Mouse ID"].isin(["g989"]) == False].reset_index()
cleaned_df.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

drug_regimen_tumor_vol = cleaned_df.loc[:, ["Drug Regimen", "Tumor Volume (mm3)"]]

mean_tumor_vol = drug_regimen_tumor_vol.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].mean()
medi_tumor_vol = drug_regimen_tumor_vol.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].median()
var_tumor_vol = drug_regimen_tumor_vol.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].var()
std_tumor_vol = drug_regimen_tumor_vol.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].std()
sem_tumor_vol = drug_regimen_tumor_vol.groupby(["Drug Regimen"])["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.

tumor_summary_table = pd.DataFrame({"Tumor Volume Mean": mean_tumor_vol,
                                "Tumor Volume Median": medi_tumor_vol,
                                "Tumor Volume Variance": var_tumor_vol,
                                "Tumor Volume Standard Deviation": std_tumor_vol,
                                "Tumor Volume SEM": sem_tumor_vol})
tumor_summary_table["Tumor Volume Mean"] = tumor_summary_table["Tumor Volume Mean"]
tumor_summary_table["Tumor Volume Median"] = tumor_summary_table["Tumor Volume Median"]
tumor_summary_table["Tumor Volume Variance"] = tumor_summary_table["Tumor Volume Variance"]
tumor_summary_table["Tumor Volume Standard Deviation"] = tumor_summary_table["Tumor Volume Standard Deviation"]
tumor_summary_table["Tumor Volume SEM"] = tumor_summary_table["Tumor Volume SEM"]

tumor_summary_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

cleaned_df.groupby(["Drug Regimen"]).agg({'Tumor Volume (mm3)': ['mean','median','std','sem']})
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
final_tumor_volume = cleaned_df.groupby(["Mouse ID"]).max().reset_index()
# print(final_tumor_volume)

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
last_timepoint_merged = final_tumor_volume[['Mouse ID','Timepoint']].merge(cleaned_df,on=['Mouse ID','Timepoint'], how= 'left')
last_timepoint_merged
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)

tumor_vol_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for Drugs in treatments:    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    
    tumor_vol_data = last_timepoint_merged.loc[last_timepoint_merged["Drug Regimen"]== Drugs, 'Tumor Volume (mm3)']
    # add subset 
    
    tumor_vol_list.append(tumor_vol_data)
    # Determine outliers using upper and lower bounds
    
    quartiles = tumor_vol_data.quantile([.25,.5,.75])
    lowq = quartiles[0.25]
    highq = quartiles[0.75]
    iqr = highq - lowq
    lower_bound = lowq - (1.5*iqr)
    upper_bound = highq + (1.5*iqr)
    
    print(f"{Drugs}'s potential outliers: {tumor_vol_data.loc[(tumor_vol_data < lowq)|(tumor_vol_data > highq)]}")
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumor_vol_list,labels=['Capomulin','Ramicane','Infubinol','Ceftamin'])
plt.title("Final Tumor Volume")
plt.xlabel("Drug Regimens")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

correlation_df = st.pearsonr(mouse_weight, average_tumor_volume)[0]
print("The calculated correlation coefficient = ", correlation_df )
```

```python
#Plotting linear regression
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(mouse_weight, average_tumor_volume)

weight_vs_volume = mouse_weight*slope + intercept

# Create the x and y variables for our plot 
average_tumor_volume = Capomulin_df.groupby(["Mouse ID"])["Tumor Volume (mm3)"].mean()
mouse_weight = Capomulin_df.groupby(["Mouse ID"])["Weight (g)"].mean()

# Create title and labels for our plot
plt.plot(mouse_weight,weight_vs_volume,"r-")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.title("Average Tumor Volume vs. Mouse Weight (Capomulin)")
plt.scatter(mouse_weight,average_tumor_volume)
plt.show()
```
